```python
import numpy as np
import matplotlib.pyplot as plt
```

We set a seed for reproducability.

```python
np.random.seed(123456)
```

A general rule of thumb for the bias-variance tradeoff is the following:
>The more expressive the model is, the higher the variance and the lower the bias, and vice versa.

Additionally, variance is often associated to overfitting.
While these two statements are true in general, we will see in this exercise a counterexample with $0$ bias, $0$ variance, and overfitting.

We consider a $1$-dimensional regression task where the data is generated by the function $f:x\in[0,1]\mapsto\frac{1}{1 - x^2}$. We assume this function is unknown, and try to estimate it purely from samples.

```python
def f(x):
    return 1 / (1 - x**2)
```

## Question 1
Generate $N = 5$ samples from $f$.
Compute the solution of the unregularized least-squares problem with basis functions $\phi_k(x) = x^k$ with $k\in\{0, 1, 2\}$.

```python
N = 5

xs_train = np.random.rand(N)
ys_train = f(xs_train)

def linreg(phi, y):
    A = phi.T @ phi
    b = phi @ y
    w = np.linalg.solve(A, b)
    return w
```

```python
N = 25
L = 100

# Build training sets
xs_train = np.tile(np.linspace(0,1,N).reshape([-1,1]), [1, L])
ys_train = np.sin(2*np.pi*xs_train) + np.random.normal(loc=0, scale=0.3, size=(N,L))

# Plot the first dataset
xs_plot = np.linspace(0,1,5000)
ys_plot = np.sin(2*np.pi*xs_plot)
plt.plot(xs_plot, ys_plot, 'g')
plt.scatter(xs_train[:,0], ys_train[:,0], s=80, facecolors='none', edgecolor='blue')
_ = plt.title('Ground truth and dataset with 25 samples')
```

We now run regularized linear regression with Gaussian basis functions on each of the $L=100$ datasets.

```python
def gaussian(x, mu, s):
    return np.exp(- (x-mu) **2 / (2 * s**2))
```

In the following, $M=24$ of these basis functions are used with $\mu$ chosen equidistantly in $[0,1]$ and $s=0.05$,
together with a constant bias function.

```python
M = 25
s = 0.05
mus = np.linspace(0,1,M-1)
Phi = np.hstack([xs_train[:,0].reshape([-1,1]), gaussian(xs_train[:,0].reshape([-1,1]), mus.reshape([1,-1]), s)])
```

For each of the $L$ datasets a regularized linear regression is now run with regularization parameter $\lambda$ equistant (in log-space) between -2.5 and 1.5.

```python
lambdas = np.exp(np.linspace(-3, 3, 100))
n_lambda = lambdas.size
coeffs = np.zeros([n_lambda, L, M])
for i in range(n_lambda):
    coeffs[i,:,:] = la.solve(Phi.T @ Phi + lambdas[i]*np.identity(N), Phi.T @ ys_train).T
```

We now select three different values of the regularization parameter $\lambda$ and 
plot fits from 20 randomly selected datasets (on the right left) as well as the average fit 
(i.e. the mean over all 100 fits) on the right (in red) together with the ground truth (in green).

```python
Phi_plot = np.hstack([xs_plot.reshape([-1,1]), gaussian(xs_plot.reshape([-1,1]), mus.reshape([1,-1]), s)])
plt.figure(figsize=(10,10))
n_plot = 20
    
inds_lambdas = [95, 50, 5]
for i_lambda in range(len(inds_lambdas)):
    # Select n_plot of the data sets and plot their fits for the current regularization parameter
    plt.subplot(3,2,2*i_lambda+1)
    inds_plot = np.random.choice(np.arange(L), n_plot, replace=False)
    for i_plot in range(20):
        plt.plot(xs_plot, 
                 Phi_plot @ coeffs[inds_lambdas[i_lambda], inds_plot[i_plot], :], 
                 'r',
                 linewidth=0.4)
    plt.ylim([-1.5,1.5])
    plt.xticks([0,1])
    plt.yticks([-1,0,1])
    plt.title(f'$\lambda={np.log(lambdas[inds_lambdas[i_lambda]]):0.2f}$')
    
    # Compare the average fit over all L datasets with the ground truth
    plt.subplot(3,2,2*i_lambda+2)
    plt.plot(xs_plot, ys_plot, 'g')
    plt.plot(xs_plot, np.mean(Phi_plot @ coeffs[inds_lambdas[i_lambda],:,:].T, axis=1), 'r')
    plt.ylim([-1.5,1.5])
    plt.xticks([0,1])
    plt.yticks([-1,0,1])
```

---

To further investigate what is going on, we now compute and plot the empirical squared bias, variance, their sum
and the test error (on a data set of 1000 equidstant points from $[0,1]$) for all regularization parameters.
Note that we follow Bishop, Section 3.2, and evaluate the bias and variance terms over the $N=25$ equidistant inputs.

```python
n_test = 1000
xs_test = np.linspace(0,1,n_test)
ys_test = np.sin(2*np.pi*xs_test)
Phi_test = np.hstack([xs_test.reshape([-1,1]), gaussian(xs_test.reshape([-1,1]), mus.reshape([1,-1]), s)])

mse_test_error = np.zeros(n_lambda)
squared_bias = np.zeros(n_lambda)
variances = np.zeros(n_lambda)
for i in range(n_lambda):
    mse_test_error[i] = np.sqrt(la.norm(Phi_test @ coeffs[i,:,:].T - ys_test.reshape([-1,1]), ord=2)**2/(L*n_test))
    squared_bias[i] = la.norm(np.mean(Phi @ coeffs[i,:,:].T, axis=1) - np.sin(2*np.pi*xs_train[:,0]), ord=2)**2/N
    variances[i] =  la.norm(Phi @ coeffs[i,:,:].T - np.tile(np.mean(Phi @ coeffs[i,:,:].T, axis=1).reshape([-1,1]), [1,L]), ord=2)**2/(N*L)
```

```python
plot_range = slice(0,70)
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
plt.plot(np.log(lambdas)[plot_range], squared_bias[plot_range], 'b')
plt.plot(np.log(lambdas)[plot_range], variances[plot_range], 'r')
plt.plot(np.log(lambdas)[plot_range], squared_bias[plot_range] + variances[plot_range], 'm')
_ = plt.legend(["$($bias$)^2$", "variance", "$($bias$)^2 +$ variance"])
plt.grid()

plt.subplot(1,2,2)
plt.plot(np.log(lambdas)[plot_range], mse_test_error[plot_range], 'k')
_ = plt.legend(["test error"])
plt.grid()
```
