A general rule of thumb for the bias-variance tradeoff is the following:
>The more expressive the model is, the higher the variance and the lower the bias, and vice versa.

In this exercise, we will make that statement more precise. In particular, we will see what is meant by "expressive".

We consider a $1$-dimensional regression task where the data is generated by the function $f:x\in[0,1]\mapsto\frac{1}{1 - x}$. We assume this function is unknown, and try to estimate it purely from samples.

```python
def f(x, noise=False):
    values = 1 / (1 - x)
    if noise:
        return values + np.random.normal(loc=0, scale=0.3, size=x.shape)
    else:
        return values
```

## Question 1
Compute the solution of the unregularized least-squares problem with basis functions $\phi_k(x) = x^k$ with $k\in\{0, 1, 2\}$ on $N=5$ data points.

```python
def linreg(phi, y):
    A = phi.T @ phi
    b = phi.T @ y
    w = np.linalg.inv(A) @ b
    return w

def get_polynomial_features(x, M):
    return np.vstack([
        x**k for k in range(M+1)
    ]).T

x_train_sample = np.sort(np.random.rand(5))
y_train_sample = f(x_train_sample)

M = 2

phi = get_polynomial_features(x_train_sample, M)

w = linreg(phi, y_train_sample)
x_plot = np.linspace(0, 1, 1000)
plt.figure(figsize=(15,10))
plt.plot(x_plot, get_polynomial_features(x_plot, M)@w)
_=plt.scatter(x_train_sample, y_train_sample)
print(x_train_sample, y_train_sample)
```

## Question 2

Implement and run linear regression on $N=500$ data points with basis functions $\phi_{k}(x) = x^k$, for $k\in\{0,\dots,M\}$, with $M\in\{5, 10, 15, 20\}$. Plot your prediction on $[0, 1]$ as well as the training data.
Does your prediction overfit? Underfit?

```python
def run_plot_linreg(x, y, M):
    phi = get_polynomial_features(x, M)
    w = linreg(phi, y)
    phi_plot = get_polynomial_features(x_plot, M)
    y_plot = phi_plot @ w
    plt.plot(x_plot, y_plot, label=f'M={M}')
    return w
    
x_train = np.sort(np.random.rand(500))
y_train = f(x_train)
x_plot = np.linspace(0, 1, 1000)

Ms = np.arange(5, 21, 5)
plt.figure(figsize=(15,10))
plt.ylim(-50, 200)
for k in range(Ms.shape[0]):
    M = Ms[k]
    run_plot_linreg(x_train, y_train, M)
plt.legend()
plt.scatter(x_train, y_train)

    
```

## Question 3

Run your algorithm $25$ times on different datasets with $M=5$, and plot the different functions you get on the interval $[0, 1]$ after subtracting the ground truth. Do you seem to have high variance? 
Plot the average function you get minus the ground truth. Do you seem to have a high bias?

```python
T = 50
N = 500
M = 5

# Build training sets
x_train = np.array([np.sort(np.random.rand(N)) for k in range(T)])
y_train = f(x_train)

# Do the linear regressions
def run_plot_linreg(x, y, M, x_plot, y_true):
    phi = get_polynomial_features(x, M)
    phi_plot = get_polynomial_features(x_plot, M)
    w = linreg(phi, y)
    y_plot = phi_plot @ w
    plt.plot(x_plot, y_plot - y_true, color='r',linewidth=0.4)
    return w, y_plot
    

plt.figure(figsize=(15,10))
plt.ylim(-50, 200)
w = np.zeros((M+1, T), dtype=float)
x_plot = np.linspace(0,1,1000,endpoint=False)
y_true = f(x_plot)
y_plot = np.zeros((T, x_plot.shape[0]))
for t in range(T):
    w[:, t], y_plot[t, :] = run_plot_linreg(x_train[t, :], y_train[t, :], M, x_plot, y_true)

plt.figure(figsize=(15,10))
plt.ylim(-200, 200)
plt.plot(x_plot, y_plot.mean(axis=0)-y_true, 'r')
```

**There seems to be both a very high variance and high bias**

## Question 4

We now integrate the true function $f$ in the basis functions family, and re-do the previous computations. What do you observe?

```python
T = 50
N = 500
M = 5

# Build training sets
x_train = np.array([np.sort(np.random.rand(N)) for k in range(T)])
y_train = f(x_train)

# Do the linear regressions
def get_extended_features(x, M):
    phi = get_polynomial_features(x, M)
    new_feature = f(x).reshape(-1, 1)
    phi = np.hstack((phi, new_feature))
    return phi

def run_plot_linreg(x, y, M, x_plot, y_true):
    phi = get_extended_features(x, M)
    phi_plot = get_extended_features(x_plot, M)
    w = linreg(phi, y)
    y_plot = phi_plot @ w
    plt.plot(x_plot, y_plot-y_true, color='r',linewidth=0.4)
    return w, y_plot
    

plt.figure(figsize=(15,10))
plt.ylim(-2, 2)
w = np.zeros((M+2, T), dtype=float)
x_plot = np.linspace(0,1,1000,endpoint=False)
y_true = f(x_plot)
y_plot = np.zeros((T, x_plot.shape[0]))
for t in range(T):
    w[:, t], y_plot[t, :] = run_plot_linreg(x_train[t, :], y_train[t, :], M, x_plot, y_true)

plt.figure(figsize=(15,10))
plt.ylim(-2, 2)
plt.plot(x_plot, y_plot.mean(axis=0)-y_true, 'r')
```

**There is exactly $0$ bias and $0$ variance.**

## Question 5

We now add some noise to the data. Re-do questions 2 and 3. What do you observe?

```python
T = 50
N = 500
M = 5

# Build training sets
x_train = np.array([np.sort(np.random.rand(N)) for k in range(T)])
y_train = f(x_train, noise=True)

# Do the linear regressions
def get_extended_features(x, M):
    phi = get_polynomial_features(x, M)
    new_feature = f(x).reshape(-1, 1)
    phi = np.hstack((phi, new_feature))
    return phi

def run_plot_linreg(x, y, M, x_plot, y_true):
    phi = get_extended_features(x, M)
    phi_plot = get_extended_features(x_plot, M)
    w = linreg(phi, y)
    y_plot = phi_plot @ w
    plt.plot(x_plot, y_plot-y_true, color='r',linewidth=0.4)
    return w, y_plot
    

plt.figure(figsize=(15,10))
plt.ylim(-2, 2)
w = np.zeros((M+2, T), dtype=float)
x_plot = np.linspace(0,1,1000,endpoint=False)
y_true = f(x_plot)
y_plot = np.zeros((T, x_plot.shape[0]))
for t in range(T):
    w[:, t], y_plot[t, :] = run_plot_linreg(x_train[t, :], y_train[t, :], M, x_plot, y_true)

plt.figure(figsize=(15,10))
plt.ylim(-2, 2)
plt.plot(x_plot, y_plot.mean(axis=0)-y_true, 'r')
```

**There is still no bias: the basis function $f$ perfectly captures the true variation of the data. However, some variance is now introduced because of the noise. 
But we know that this is pure noise: the polynomials overfit to the noise!**

A better intuitive definition of bias and variance:
> Bias: Measures how bad your hypothesis class is to approximate the true function



> Variance: Measures how much your approximation depends on the individual dataset

Here, polynomials are a very poor approximation of $f$.
In general, variance is characterized by wildly varying predictions
