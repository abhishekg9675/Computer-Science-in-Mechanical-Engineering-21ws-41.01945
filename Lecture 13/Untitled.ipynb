## Simple Bayesian Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Data
x_train = np.linspace(-5,5,9).reshape((-1,1))
y_train = 7.6 * x_train + np.random.normal(0,2, np.shape(x_train))

x_test = np.array([-4.5, -2.5, -0.5, 1., 2.8]).reshape((-1,1))
y_test = 7.6 * x_test
```

```python
plt.plot(x_train, y_train, color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=12)

plt.xlabel('x_train')
plt.ylabel('y_train')
 
plt.show()
```

```python
#Assume Prior model
w_mean_prior, w_var_prior = 0, 1
#w_prior = np.random.normal(w_mean_prior, w_var_prior, (1,1))


#Plotting prior model
ax1 = plt.subplots(1,1)

x_axis = np.linspace(-5,5, 1000)
y_axis = np.random.normal(w_mean_prior, np.sqrt(w_var_prior), 100)
plt.plot(x_axis, stats.norm.pdf(x_axis, w_mean_prior, np.sqrt(w_var_prior)), 'black', label='PDF')
plt.plot([w_mean_prior, w_mean_prior],[0, stats.norm.pdf(w_mean_prior, w_mean_prior, np.sqrt(w_var_prior))],'tab:orange', label='mean') 
ws_std = np.linspace(-1*1,1*1,1000)
ws_std_pdf = stats.norm.pdf(ws_std, w_mean_prior, np.sqrt(w_var_prior))
plt.fill_between(ws_std,0,ws_std_pdf,facecolor='lightgrey',label='+/-1 std')
plt.xlabel('w')
plt.ylabel('p(w)')
plt.title('Prior distribution of weights')
plt.legend()

ax1 = plt.subplots(1,1)

xs = np.linspace(-5,5,1000)
plt.plot(xs,w_mean_prior*xs,'tab:orange')
plt.fill_between(xs,1*np.sqrt(w_var_prior)*xs,-1*1*np.sqrt(w_var_prior)*xs,facecolor='lightgrey')
plt.xlabel('x')
plt.ylabel('y=w . x')
plt.title('Distribution over functions')
```
